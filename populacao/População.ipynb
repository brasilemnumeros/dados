```python
import ipeadatapy as ipea
```

```python
# Carregar os dados da série 'DEPIS_POP' usando ipeadatapy
populacao_df = ipea.timeseries('DEPIS_POP')

# Exibir as primeiras linhas do DataFrame carregado
display(populacao_df.head())

display(populacao_df.tail())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assegurar que o índice do DataFrame seja do tipo datetime para plotagem
populacao_df.index = pd.to_datetime(populacao_df.index)

# Plotar o gráfico de linha da população ao longo do tempo
plt.figure(figsize=(12, 6))
plt.plot(populacao_df.index, populacao_df['VALUE (Habitante)'])
plt.title('População do Brasil ao Longo do Tempo')
plt.xlabel('Ano')
plt.ylabel('População (Habitante)')
plt.grid(True)
plt.show()
```

```python
# Filtrar o DataFrame populacao_df a partir do ano de 1995
populacao_filtrada_df = populacao_df[populacao_df.index.year >= 1995].copy()

# Exibir as primeiras linhas do DataFrame filtrado
display(populacao_filtrada_df.head())

# Exibir as últimas linhas para confirmar o filtro
display(populacao_filtrada_df.tail())
```

```python
import matplotlib.pyplot as plt

# Plotar o gráfico de linha da população filtrada (a partir de 1995)
plt.figure(figsize=(12, 6))
plt.plot(populacao_filtrada_df.index, populacao_filtrada_df['VALUE (Habitante)'])
plt.title('População do Brasil a partir de 1995')
plt.xlabel('Ano')
plt.ylabel('População (Habitante)')
plt.grid(True)
plt.show()
```

```python
# Calcular a variação percentual ano a ano para o DataFrame filtrado
populacao_filtrada_df['Variacao Anual (%) Recorte'] = populacao_filtrada_df['VALUE (Habitante)'].pct_change() * 100

# Exibir as primeiras linhas do DataFrame filtrado com a nova coluna de variação percentual do recorte
display(populacao_filtrada_df.head())

# Plotar o gráfico de barras da variação anual percentual da população a partir de 1995
plt.figure(figsize=(15, 7))
bars = plt.bar(populacao_filtrada_df.index, populacao_filtrada_df['Variacao Anual (%) Recorte'], color='skyblue', width=0.8)

plt.title('Variação Anual Percentual da População do Brasil a partir de 1995')
plt.xlabel('Ano')
plt.ylabel('Variação Populacional (%)')
plt.grid(axis='y')

# Opcional: adicionar valores acima das barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.2f}%', va='bottom')

plt.show()
```

```python
import json
import os

# Definir o diretório para salvar os arquivos JSON
output_dir = "/content"
os.makedirs(output_dir, exist_ok=True)

# Função para formatar e salvar dados em JSON
def save_to_json(dataframe, value_column, filename):
    data_list = []
    for index, row in dataframe.iterrows():
        # Formatar a data como YYYY-MM-DD
        date_str = index.strftime('%Y-%m-%d')
        data_list.append({"date": date_str, "value": row[value_column]})

    # Criar a estrutura JSON final
    json_output = {"data": data_list}

    # Definir o caminho completo do arquivo
    filepath = os.path.join(output_dir, filename)

    # Salvar em arquivo JSON
    with open(filepath, 'w', encoding='utf-8') as f:
        json.dump(json_output, f, ensure_ascii=False, indent=4)

    print(f"Dados salvos em: {filepath}")

# Salvar população total (DataFrame filtrado a partir de 1995)
save_to_json(populacao_filtrada_df, 'VALUE (Habitante)', 'populacao-total-anual.json')

# Salvar variação anual percentual (DataFrame filtrado a partir de 1995)
# Remover o primeiro valor NaN antes de salvar
variacao_anual_percentual_filtrada_df = populacao_filtrada_df.dropna(subset=['Variacao Anual (%) Recorte']).copy()
save_to_json(variacao_anual_percentual_filtrada_df, 'Variacao Anual (%) Recorte', 'populacao-variacao-anual.json')
```
